```python
import pysolar as ps
import datetime # to generate datetime objects
import pytz # generate a timezone object
import matplotlib.pyplot as plt # performing plots
```

```python
queens_lat = 40.742054
queens_long = -73.769417
```

```python
est = pytz.timezone('America/New_York')
date = datetime.datetime(2021, 1, 1,12,0,0,0,est)
```

```python
print(ps.solar.get_altitude(queens_lat, queens_long, date))
print(ps.solar.get_azimuth(queens_lat, queens_long, date))
```

```python
azi = []
alti = []
ana_date = datetime.datetime(2021, 1, 1,12,0,0,0,est) # First date of the year

for day in range(0,364):
    ana_date = ana_date + datetime.timedelta(days = 1) # increment by a single day
    altitude = ps.solar.get_altitude(queens_lat, queens_long, ana_date)
    azimuth = ps.solar.get_azimuth(queens_lat, queens_long, ana_date)
    azi.append(azimuth)
    alti.append(altitude)
plt.title ("Sun Analemma Plot in 2021 (12pm) Queens, NYC, USA") 
plt.xlabel("Azimuth (°)")   
plt.ylabel("Altitude (°)")
plt.scatter(azi, alti)
plt.rcParams['figure.dpi'] = 500
plt.savefig('"Sun Analemma Plot in 2021 (12pm) Sydney, Australia.png', format = 'png', dpi=500)
```
